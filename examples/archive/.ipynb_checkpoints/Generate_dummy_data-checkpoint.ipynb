# The art of data mimicry
Mimic original distributions: by random grabbing from all previous values 

```python
import pandas as pd

# Generate faker for 4 different clusters

# Variables from development set
df_numeric_raw = pd.read_csv('../../RA_Clustering/new_data/5_clustering/df_lab_raw.csv', sep=',')[['Leuko', 'MCH', 'Hb', 'Ht', 'MCHC',  'MCV', 'Trom', 'BSE', 'Lym', 'Mono']] #, 'Age' 
df_categoric_raw= pd.read_csv('../example_data/development/ClusteringDEV_CATEGORICAL.csv', sep=',')
```

## Strategy: random grab & pollute
ToDo: generating the original distribution for each cluster is probably more efficient? 



```python
import numpy as np

N_PAT = 100
```

### # Numeric

```python
df_dummy_num = pd.DataFrame()

for col in df_numeric_raw.columns: 
    max_deviation = df_numeric_raw[col].std()/2
    noise = np.random.uniform(low=-max_deviation, high=max_deviation, size=N_PAT)
    noise_2 = np.random.uniform(low=-max_deviation, high=max_deviation, size=N_PAT)
    
    # Grab random existing value
    val = np.random.choice(df_numeric_raw[col].values, N_PAT)
    
    # Add noise to value 
    df_dummy_num[col] = np.round(np.abs(val + noise + noise_2), 3)

# export
df_dummy_num.to_csv(r'../dummy_data/NumericalModality_dummy.csv', sep=';', index=False)
df_dummy_num.head()
```

### # Categoric

```python
df_dummy_cat = pd.DataFrame()

# df_categoric_raw.columns[1:] #
l_cat_cols = [col for col in df_categoric_raw.columns if ('_positive' in col or col in ['Sex', 'RF', 'aCCP']) ]

for col in l_cat_cols: 
    val = np.random.choice([0,1], N_PAT, p=[round(len(df_categoric_raw[df_categoric_raw[col]==0])/len(df_categoric_raw),2), round(len(df_categoric_raw[df_categoric_raw[col]==1])/len(df_categoric_raw),2)])
    
    # Add value to df dummy 
    df_dummy_cat[col] = val
    
    # add negative version
    if 'positive' in col:
        df_dummy_cat[col.split('_positive')[0] + '_negative'] = np.abs(1 - val)

# export
df_dummy_cat.to_csv(r'../dummy_data/CategoricalModality_dummy.csv', sep=';', index=False)
df_dummy_cat.head()
```
